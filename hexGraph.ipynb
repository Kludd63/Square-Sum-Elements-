```python
def find3path(n):
    sqrs = {}
    
    if n.is_square() == True:
        sqrs[(sqrt(n)-1)^2] = 'c'
        sqrs[n] = 'a'
        sqrs[(sqrt(n)+1)^2] = 'b'
        
    else:
        sqrs[isqrt(n)^2] = 'c'
        sqrs[(isqrt(n)+1)^2] = 'a'
        sqrs[(isqrt(n)+2)^2] = 'b'
        
    g = {}
    es = 0
    for s in sqrs.keys():
        if s%2 == 0:
            es = s
     
    for m in range(1,n+1):
        temp = dict()
        for s in sqrs.keys():
            if s-m <=n and s-m >0 and s-m !=m:
                temp[s-m] = sqrs[s]
        g[m] = temp

    ch = True
    ct = 0
    t = 0
    for h in g.keys():
        if len(g[h]) <2:
            ct +=1
                                
    if ct > 2:
        ch = False

    if ch == True:
        G= Graph(g)
        posi = dict()
        posi[es//2] = (1, 1)

        #G.plot(iterations = 80).show() 
        HP = G.hamiltonian_path()
        IG = list(G.breadth_first_search([1]))
        if HP != None:
            m = 0
            for v in IG:
                a = 0
                b = 0
                c = 0
                p = G.shortest_path(es//2, v)
                for i in range(1, len(p)):
                    if sqrs[p[i]+p[i-1]] == 'a':  
                        a += 1
                    elif sqrs[p[i]+p[i-1]] == 'b':
                        b += 1
                    elif sqrs[p[i]+p[i-1]] == 'c':
                        c += 1 
                if a == 0 and b+c >1:
                    if m == 0:
                        posi[v] = (1, 1 +(b+c))
                        m +=1 
                    else:
                        posi[v] = (1, 1 - (b+c))
                else:
                    posi[v] = (posi[es//2][0] + (a*1) + (b*1) + (c*1), posi[es//2][1] + (b*1) + (c*-1))
                
            
            G.plot(pos = posi).show()
            tec = Graph(g, pos = posi)
            tec.set_latex_options(tkz_style = "Classic", graphic_size = (15,2), vertex_size = 0.4, edge_thickness = 0.3, vertex_label_placement = (.4, 270))
            print(latex(tec))
            return f"Path found for {n} using {sqrs}, creating graph {g}."
    return f"No path found for {n} using only 3 squares"
```
